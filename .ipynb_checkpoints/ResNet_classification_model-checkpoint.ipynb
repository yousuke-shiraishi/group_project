### データセットを準備

```python
#分類するクラスを設定する
classes = ['coffee_img', 'water_img']
nb_classes = len(classes)
 
# 学習データと検証データのファイル場所を設定する
train_data_dir = './drive/My Drive/Colab Notebooks/sprint26/video_frame_img/train'
validation_data_dir = './drive/My Drive/Colab Notebooks/sprint26/video_frame_img/valid'
 
# 学習データと検証データの数を設定
nb_train_samples = 2602
nb_validation_samples = 20

# イメージサイズを設定する
img_width, img_height = 224, 224
```

### オーグメンテーションを実施する場合のジェネレータ

```python
from keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(rescale=1.0 / 255, zoom_range=0.2, horizontal_flip=True)
validation_datagen = ImageDataGenerator(rescale=1.0 / 255)
```

```python
train_generator = train_datagen.flow_from_directory(
  train_data_dir,
  target_size=(img_width, img_height),
  color_mode='rgb',
  classes=classes,
  class_mode='categorical',
  batch_size=16)
 
validation_generator = validation_datagen.flow_from_directory(
  validation_data_dir,
  target_size=(img_width, img_height),
  color_mode='rgb',
  classes=classes,
  class_mode='categorical',
  batch_size=16)
```

# 転移学習モデルの作成

```python
from keras.applications.resnet50 import ResNet50
from keras.models import Sequential, Model
from keras.layers import Input, Flatten, Dense
```

### データ型の調整

```python
# インプットデータ型を設定
input_tensor = Input(shape=(img_width, img_height, 3))
# ResNetを活用
ResNet50 = ResNet50(include_top=False, weights='imagenet',input_tensor=input_tensor)
```

### モデルの作成

```python
top_model = Sequential()
top_model.add(Flatten(input_shape=ResNet50.output_shape[1:]))
top_model.add(Dense(nb_classes, activation='sigmoid'))
```

```python
	
model = Model(input=ResNet50.input, output=top_model(ResNet50.output))
```

### モデルのコンパイル

```python
from keras import optimizers
model.compile(loss='categorical_crossentropy',
              optimizer=optimizers.SGD(lr=1e-3, momentum=0.9),
              metrics=['accuracy'])
```

### モデルの学習

```python
history = model.fit_generator(
    train_generator,
    samples_per_epoch=nb_train_samples,
    nb_epoch=1,
    validation_data=validation_generator,
    nb_val_samples=nb_validation_samples)
```

```python
from keras.preprocessing import image
import numpy as np

 
image_data = './drive/My Drive/Colab Notebooks/sprint26/video_frame_img/test/coffee_img/katemita-2018-01-24-suntory-craftboss-01.jpg'
img = image.load_img(image_data, target_size=(img_width, img_height))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = x / 255
pred = model.predict(x)[0]

 
# 予測確率を出力
top = 2
top_indices = pred.argsort()[-top:][::-1]
result = [(classes[i], pred[i]) for i in top_indices]
print(result)
```
